# 🌿 Jaringan Semantik - Ekosistem & Lingkungan
Implementasi Semantic Network berbasis Python dengan tema Ekosistem dan Lingkungan.

**Relasi yang digunakan:**
- `is-a` : jenis/kategori
- `bagian-dari` : merupakan bagian dari
- `menyerap` : menyerap sesuatu
- `menghasilkan` : menghasilkan sesuatu
- `tempat-hidup` : tempat tinggal makhluk hidup
- `ancaman` : ancaman terhadap ekosistem
- `fungsi` : fungsi/peran dalam ekosistem

```python
# ============================================
# GRAF JARINGAN SEMANTIK - EKOSISTEM
# ============================================

graph = {
    # --- TUMBUHAN ---
    "pohon": {
        "is-a": "tumbuhan",
        "bagian-dari": "hutan",
        "menyerap": "CO2",
        "menghasilkan": "oksigen"
    },
    "rumput": {
        "is-a": "tumbuhan",
        "bagian-dari": "padang rumput",
        "fungsi": "penutup tanah"
    },
    "mangrove": {
        "is-a": "tumbuhan",
        "bagian-dari": "pesisir",
        "fungsi": "pelindung pantai",
        "menyerap": "CO2"
    },
    "tumbuhan": {
        "is-a": "makhluk hidup",
        "fungsi": "produsen ekosistem",
        "menyerap": "CO2"
    },

    # --- HEWAN ---
    "harimau": {
        "is-a": "hewan",
        "tempat-hidup": "hutan",
        "fungsi": "predator puncak"
    },
    "lebah": {
        "is-a": "hewan",
        "tempat-hidup": "hutan",
        "fungsi": "penyerbuk tanaman"
    },
    "ikan": {
        "is-a": "hewan",
        "tempat-hidup": "sungai",
        "fungsi": "penjaga rantai makanan air"
    },
    "hewan": {
        "is-a": "makhluk hidup",
        "fungsi": "konsumen ekosistem"
    },

    # --- EKOSISTEM ---
    "hutan": {
        "is-a": "ekosistem",
        "menyerap": "CO2",
        "menghasilkan": "oksigen",
        "ancaman": "deforestasi"
    },
    "sungai": {
        "is-a": "ekosistem",
        "fungsi": "sumber air bersih",
        "ancaman": "pencemaran"
    },
    "padang rumput": {
        "is-a": "ekosistem",
        "fungsi": "habitat hewan herbivora",
        "ancaman": "alih fungsi lahan"
    },
    "pesisir": {
        "is-a": "ekosistem",
        "fungsi": "pelindung daratan dari abrasi",
        "ancaman": "pembuangan limbah"
    },
    "ekosistem": {
        "is-a": "lingkungan hidup",
        "fungsi": "menjaga keseimbangan alam"
    },

    # --- MAKHLUK HIDUP ---
    "makhluk hidup": {
        "bagian-dari": "ekosistem",
        "fungsi": "komponen biotik"
    }
}
```

```python
# ============================================
# FUNGSI PENCARIAN PROPERTI (dengan pewarisan)
# ============================================

def get_property(node, prop, visited=None):
    if visited is None:
        visited = set()

    # Hindari loop tak terbatas
    if node in visited:
        return "tidak diketahui"
    visited.add(node)

    # Cek apakah node ada di graph
    if node not in graph:
        return "objek tidak ditemukan dalam jaringan"

    # Jika properti ditemukan langsung
    if prop in graph[node]:
        return graph[node][prop]

    # Jika tidak ada, naik ke parent via is-a (inheritance)
    if "is-a" in graph[node]:
        parent = graph[node]["is-a"]
        return get_property(parent, prop, visited)

    return "tidak diketahui"
```

```python
# ============================================
# FUNGSI MEMPROSES PERTANYAAN
# ============================================

def jawab_pertanyaan(kalimat):
    kata = kalimat.lower().strip().split()

    try:
        # Tangani objek dua kata (misal: "padang rumput")
        if len(kata) >= 2 and " ".join(kata[:2]) in graph:
            objek = " ".join(kata[:2])
            sisa = kata[2:]
        else:
            objek = kata[0]
            sisa = kata[1:]

        # Format: "<objek> adalah"
        if "adalah" in sisa:
            hasil = get_property(objek, "is-a")
            return f"{objek} adalah {hasil}"

        # Format: "<objek> menyerap apa"
        elif "menyerap" in sisa:
            hasil = get_property(objek, "menyerap")
            return f"{objek} menyerap {hasil}"

        # Format: "<objek> menghasilkan apa"
        elif "menghasilkan" in sisa:
            hasil = get_property(objek, "menghasilkan")
            return f"{objek} menghasilkan {hasil}"

        # Format: "<objek> bagian dari apa"
        elif "bagian" in sisa:
            hasil = get_property(objek, "bagian-dari")
            return f"{objek} adalah bagian dari {hasil}"

        # Format: "<objek> tempat hidup"
        elif "tempat" in sisa or "hidup" in sisa:
            hasil = get_property(objek, "tempat-hidup")
            return f"{objek} hidup di {hasil}"

        # Format: "<objek> ancaman"
        elif "ancaman" in sisa:
            hasil = get_property(objek, "ancaman")
            return f"ancaman terhadap {objek} adalah {hasil}"

        # Format: "<objek> fungsi"
        elif "fungsi" in sisa:
            hasil = get_property(objek, "fungsi")
            return f"fungsi {objek} adalah {hasil}"

        else:
            return (
                "Format tidak dikenali. Coba gunakan kata kunci:\n"
                "  'adalah', 'menyerap', 'menghasilkan', 'bagian',\n"
                "  'tempat hidup', 'ancaman', atau 'fungsi'\n"
                "Contoh: 'pohon menyerap', 'hutan ancaman', 'harimau fungsi'"
            )

    except Exception as e:
        return f"Error: {e}"
```

```python
# ============================================
# PROGRAM UTAMA - LOOP TANYA JAWAB INTERAKTIF
# ============================================

print("=" * 50)
print("  🌿 JARINGAN SEMANTIK EKOSISTEM & LINGKUNGAN")
print("=" * 50)
print("Objek yang tersedia:")
print(" ", ", ".join(graph.keys()))
print("\nCara bertanya:")
print("  pohon menyerap       → apa yang diserap pohon")
print("  hutan ancaman        → ancaman terhadap hutan")
print("  harimau tempat hidup → habitat harimau")
print("  mangrove adalah      → kategori mangrove")
print("  lebah fungsi         → fungsi lebah")
print("  ketik 'exit' untuk keluar")
print("=" * 50)

while True:
    tanya = input("\nPertanyaan: ")

    if tanya.lower().strip() == "exit":
        print("Terima kasih! 🌱")
        break

    if tanya.strip() == "":
        print("Input tidak boleh kosong.")
        continue

    jawaban = jawab_pertanyaan(tanya)
    print("Jawaban:", jawaban)
```
